```python
def median(list):
    x = 0
    list_s = list.sort()
    if list_s.__len__() == 0:
        return None
    elif len(list_s) % 2 == 0:
        x = len(list_s) / 2
        y = (list_s[x] + list_s[x+1])/2
        return y
    else:
        x = len(list_s) // 2
        y = list_s[x+1]
        return y

if __name__ == '__main__':
    print(median([]) is None)
    print(median([1,2,3]) == 2)
    print(median([3,7,10,4,1,9,6,5,2,8]) == 5.5)
    print(median([3,7,10,4,1,9,6,2,8]) == 6)

import random
random.seed(42)
testData = [3,7,10,4,1,9,6,2,8]
for i in range(1000):
    random.shuffle(testData)
    assert median(testData) == 6
```
